```python
def game_output(x_player: str, y_player: str) -> str:
    """return one of state draw/win/lose"""
    if x_player == y_player:
        return 'draw'
    if x_player == 'Rock':
        return 'lose' if y_player == 'Scissors' else 'win'
    if x_player == 'Paper':
        return 'win' if y_player == 'Scissors' else 'lose'
    if x_player == 'Scissors':
        return 'lose' if y_player == 'Paper' else 'win'
```

```python
def get_points_by_choice(choice: str) -> str:
    if choice == 'Rock':
        return 1
    if choice == 'Paper':
        return 2
    if choice == 'Scissors':
        return 3
```

```python
def game_points(result: str) -> str:
    if result == 'draw':
        return 3
    if result == 'win':
        return 6
    if result == 'lose':
        return 0
```

```python
opponent_map = {'A': 'Rock', 'B': 'Paper', 'C':'Scissors'}
my_options_map = {'X': 'Rock', 'Y': 'Paper', 'Z':'Scissors'}
def get_score_by_line(moves: str) -> bool:
    opponent, my_choice = moves.split('-')
    opponent_mapped = opponent_map[opponent]
    my_choice_mapped = my_options_map[my_choice]

    point_by_choice = get_points_by_choice(my_choice_mapped)

    game_result = game_output(opponent_mapped, my_choice_mapped)
    game_point = game_points(game_result)

    all_points = game_point + point_by_choice
    print(f'Allpoints: {all_points}')

    return all_points
```
